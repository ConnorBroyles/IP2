### Renaming and Copying Original Data Set

```python
df_copy=df.copy()
```

```python
df.info()
```

```python
df=df.rename(columns={'Date/Time':'date_time'})
df=df.rename(columns={'Temp (C)':'temp_c'})
df=df.rename(columns={'Dew Point Temp (C)':'dew_point'})
df=df.rename(columns={'Rel Hum (%)':'relative_humidity'})
df=df.rename(columns={'Wind Spd (km/h':'wind_speed'})
df=df.rename(columns={'Visibility (km)':'visibility'})
df=df.rename(columns={'Stn Press (kPa)':'pressure'})
df=df.rename(columns={'Weather':'weather'})
```

```python
df.info()
```

## Getting basic Stats on data

```python
df.groupby(['date_time','temp_c','dew_point','relative_humidity','visibility',
           'pressure','weather']).count()
```

```python
df.groupby(['date_time','temp_c','dew_point','relative_humidity','visibility',
           'pressure','weather']).median()
```

## Coverting Temp C to Temp in F 

```python
def f(x):
    x = x * 1.8 + 32
    return float(x)
```

```python
df['temp_f']=df['temp_c'].apply(f)
```

```python
df.info()
```
